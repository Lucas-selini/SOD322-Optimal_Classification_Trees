```python
D= [2, 3, 4]
iris = [3/30, 4/30, 0]
seeds = [0, 2/42, 7/42]
wine = [2/30, 1/30, 2/30]
dry_bean = [5/30, 6/30, 7/30]
steel_industry = [10/30, 10/30, 14/30]
```

```python
import matplotlib.pyplot as plt

iris_percentage = [value * 100 for value in iris]
seeds_percentage = [value * 100 for value in seeds]
wine_percentage = [value * 100 for value in wine]
dry_bean_percentage = [value * 100 for value in dry_bean]
steel_industry_percentage = [value * 100 for value in steel_industry]

plt.plot(D, iris_percentage, marker='o', linestyle='-', label='iris')
plt.plot(D, seeds_percentage, marker='o', linestyle='-', label='seeds')
plt.plot(D, wine_percentage, marker='o', linestyle='-', label='wine')
plt.plot(D, dry_bean_percentage, marker='o', linestyle='-', label='dry_bean')
plt.plot(D, steel_industry_percentage, marker='o', linestyle='-', label='steel_industry')

plt.xlabel('D')
plt.ylabel('Taux de mauvaise classification (%)')
plt.legend()
plt.show()
```
